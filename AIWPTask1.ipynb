```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error as mse
```

```python
df = pd.read_csv('data/Salary_Data.csv')
df.head(10)
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
```

```python
x_train, x_test, y_train, y_test = tts(x, y, random_state=0, test_size=3/10)
regr = LinearRegression()
regr.fit(x_train, y_train)
```

```python
y_pred = regr.predict(x_test)
y_pred
```

```python
regr.intercept_, regr.coef_
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_train, regr.predict(x_train), color='green')
plt.title('training set')
```

```python
plt.scatter(x_test, y_test, color='blue')
plt.plot(x_test, regr.predict(x_test), color='red')
plt.title('testing set')
```

```python
def fn_mse(y_pred, y_test): 
    return np.square(np.subtract(y_pred,y_test)).mean() 
```

```python
def fn_rmse(y_pred, y_test): 
    return np.sqrt(fn_mse(y_pred, y_test))
```

```python
def fn_rmse_2(y_pred, y_test):
    return np.sqrt(np.square(np.subtract(y_pred, y_test)).mean())
```

```python
r2_score(y_pred, y_test)
```

```python
fn_mse(y_pred, y_test), fn_rmse(y_pred, y_test),fn_rmse_2(y_pred, y_test)
```

```python
mse(y_pred, y_test)
```

```python
rmse = np.sqrt(mse(y_pred, y_test))
rmse
```
